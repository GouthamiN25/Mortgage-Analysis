# Data Visualization


```r
library(ggplot2)
```

```r
# Visualization 1: Bar chart of the most common issues
issue_counts <- new_data %>%
  group_by(issue) %>%
  summarize(count = n()) %>%
  arrange(desc(count))

ggplot(issue_counts, aes(x = reorder(issue, -count), y = count)) +
  geom_bar(stat = "identity", fill = "blue") +
  labs(title = "Common Issues",
       x = "Issue",
       y = "Count") +
 theme(axis.text.x = element_text(angle = 90, hjust = 1, margin = margin(t = 2)))
```

```r
# Visualization 2: Box plot of complaint counts by state
ggplot(new_data, aes(x = state, y = complaint_id)) +
  geom_boxplot() +
  labs(title = "Complaints vs State")+
  theme(axis.text.x = element_text(angle = 90, hjust = 1))
```

```r
# Visualization 3: Histogram of product distribution
ggplot(new_data, aes(x = product, fill = product)) +
  geom_bar() +
  labs(title = "Distribution of Products",
       x = "Product",
       y = "Count") +
  theme(axis.text.x = element_text(angle = 90, hjust = 1))
```

```r
# Visualization 4: Scatter plot of response times
new_data$date_received <- as.Date(new_data$date_received, format = "%m/%d/%Y")
new_data$date_sent_to_company <- as.Date(new_data$date_sent_to_company, format = "%m/%d/%Y")

ggplot(new_data, aes(x = date_received, y = date_sent_to_company)) +
  geom_point() +
  labs(title = "Response Time Over Time",
       x = "Date Received",
       y = "Date Sent to Company") +
  theme(axis.text.x = element_text(angle = 90, hjust = 1))
```

#Simple Linear Regression:



```r
subset_data <- new_data2[, c('zipcode', 'complaint_id')]

# Convert 'zipcode' to numeric (remove commas)
subset_data$zipcode <- as.numeric(gsub(",", "", subset_data$zipcode))


model_1<- lm(complaint_id ~ zipcode, data = subset_data)
summary(model_1)
ggplot(subset_data, aes(x = zipcode, y = complaint_id)) +
  geom_point() +
  geom_smooth(method = "lm", se = FALSE, color = "blue") +
  labs(title = "Simple Linear Regression: zipcode vs. complaint_id")
```

```r
subset_data2 <- new_data2[, c('state', 'complaint_id')]
model_2 <- lm(complaint_id ~ state, data = subset_data2)
ggplot(subset_data2, aes(x = state, y = complaint_id)) +
  geom_point() +
  geom_smooth(method = "lm", se = FALSE, color = "red") +
  labs(title = "Simple Linear Regression: state vs. complaint_id")+
   theme(axis.text.x = element_text(angle = 90, hjust = 1))
```

```r
subset_data3 <- new_data2[, c('product', 'complaint_id')]
model_3 <- lm(complaint_id ~ product, data = subset_data3)
ggplot(subset_data3, aes(x = product, y = complaint_id)) +
  geom_point() +
  geom_smooth(method = "lm", se = FALSE, color = "blue") +
  labs(title = "Simple Linear Regression: product vs. complaint_id")+
   theme(axis.text.x = element_text(angle = 90, hjust = 1))
 summary(model_3)
```

```r
subset_data4 <- new_data2[, c('company_response_to_consumer', 'complaint_id')]

model_4 <- lm(complaint_id ~ company_response_to_consumer, data = subset_data4)

ggplot(subset_data4, aes(x = company_response_to_consumer, y = complaint_id)) +
  geom_point() +
  geom_smooth(method = "lm", se = FALSE, color = "blue") +
  labs(title = "Simple Linear Regression: company_response_to_consumer vs. complaint_id")+
   theme(axis.text.x = element_text(angle = 90, hjust = 1))

 summary(model_4)
```

```r
subset_data5 <- new_data2[, c('submitted_via', 'complaint_id')]

model_5<- lm(complaint_id ~ submitted_via, data = subset_data5)

ggplot(subset_data5, aes(x = submitted_via, y = complaint_id)) +
  geom_point()+
  labs(title = "Simple Linear Regression: submitted_via vs. complaint_id")+
   theme(axis.text.x = element_text(angle = 90, hjust = 1))
 summary(model_5)
```
